```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with actual data from the image)
data = {
    'Coin': ['BTC', 'ETH', 'XRP', 'USDT', 'SOL', 'BNB', 'DOGE', 'USDC', 'ADA', 'TRX'],
    'Market Cap': [2.08e12, 398.29e9, 181.94e9, 138.35e9, 118.77e9, 99.43e9, 56e9, 48.72e9, 35.16e9, 20.83e9],
    'Perf 1W': [10.97, 5.40, 25.32, -0.02, 33.54, 0.32, 12.20, -0.03, 5.90, 8.54],
    'Perf 1M': [7.84, -0.95, 41.30, 0.00, 34.60, 4.14, 18.41, -0.06, 10.52, -1.20],
    'Perf 3M': [55.58, 26.08, 492.34, 0.02, 45.70, 16.33, 171.54, -0.05, 174.46, 50.95],
    'Perf 6M': [60.36, -0.93, 411.19, -0.04, 37.05, 20.11, 196.28, -0.06, 146.15, 79.93]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert 'Market Cap' to float for consistency
df['Market Cap'] = df['Market Cap'].astype(float)

# Sort by market cap in descending order
df = df.sort_values(by='Market Cap', ascending=False)

# Print the sorted DataFrame
print(df)

# Visualize the performance of top 5 coins over 3 months
if len(df) >= 5:
    top_5_coins = df.head(5)
    plt.figure(figsize=(10, 6))
    plt.bar(top_5_coins['Coin'], top_5_coins['Perf 3M'])
    plt.xlabel('Coin')
    plt.ylabel('3 Month Performance (%)')
    plt.title('Top 5 Coins - 3 Month Performance')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("Not enough data for visualization. Need at least 5 entries.")
```
